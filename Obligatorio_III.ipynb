```python
def calc_info_gain(df,t_feature):
    cols=list(df)
    cols.remove(t_feature)
    rem = []
    tot_rows = df.shape[0]
    init_H = entropy(df[t_feature].value_counts(),base=2)
    for c in cols:
        unique_vals = df[c].unique()
        entropy_vals = []
        for uv in unique_vals:
            df_val = df[df[c]==uv]
            row_count = df_val.shape[0]
            val_vc = df_val[t_feature].value_counts()
            entropy_vals.append(entropy(val_vc,base=2)*row_count)
        rem.append([c,sum(entropy_vals)/tot_rows])
    rem = zip(*rem)
    return zip(rem[0], init_H - rem[1])
```

```python
calc_info_gain(ej_1_df,"Cancer_Risk")
```

```python
def logistic(val):
    return 1/(1+np.exp(val))


def prob_dangerous(dose_1, dose_2):
    return logistic(-0.848 + 1.545*dose_1 - 1.942*dose_2 + 1.973*dose_1**2 + 2.495*dose_2**2 + 0.104*dose_1**3 + 0.095*dose_2**3 + 3.009*dose_1*dose_2)
```

```python
vals = [[0.5 , 0.75], [0.1, 0.75], [-0.47, -0.39], [-0.47, 0.18]]
print vals
```

```python
pred=[]
for d_1,d_2 in vals:
    pred.append(prob_dangerous(d_1,d_2))
zip(vals,pred)
```
